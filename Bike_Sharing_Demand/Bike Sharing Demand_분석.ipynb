```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

%config Completer.use_jedi = False

#노트북 안에 그래프를 그리기 위해
%matplotlib inline

#그래프에서 격자로 숫자 범위가 눈에 잘 띄도록 ggplot 스타일을 사용
plt.style.use('ggplot')

#그래프에서 마이너스 폰트 깨지는 문제에 대한 대처
mpl.rcParams['axes.unicode_minus'] = False
```

```python
# 데이터 불러오기(데이터 타입 설정)
train = pd.read_csv("data/bike-sharing-demand/train.csv", parse_dates=["datetime"])
train.shape
```

```python
#데이터 살펴보기(컬럼만 보기, 데이터 타입만 보기, 데이터에 대한 설정 다 보기)
# train.columns
# train.dtypes
train.info()
```

```python
#상위 데이터 5개 불러서 데이터 확인 -> 예측을 위해 count 데이터를 사용할 예정
train.head()
```

```python
#기온에 대해서 구체적인 정보롤 확인
train.temp.describe()
```

```python
#데이터 중 null 데이터가 있는지 보기
train.isnull().sum()
```

```python
#데이터가 없는 것에 대한 시각화 (채워져 있으면 검은색으로 표시됨)
import missingno as msno
msno.matrix(train, figsize=(12,9))
```

```python
#데이터 타입을 나누어서 연월일시분초로 데이터 프레임에 다시 담기
train['year'] = train['datetime'].dt.year
train['month'] = train['datetime'].dt.month
train['day'] = train['datetime'].dt.day
train['hour'] = train['datetime'].dt.hour
train['minute'] = train['datetime'].dt.minute
train['second'] = train['datetime'].dt.second
train.shape
```

```python
#컬럼이 세분화 되서 추가 됨 (기존 12개 -> 변경 후 18개)
train.head()
```

```python
figure, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(nrows=2, ncols=3)
figure.set_size_inches(18,8)

sns.barplot(data=train, x='year',y='count', ax=ax1)
sns.barplot(data=train, x='month',y='count', ax=ax2)
sns.barplot(data=train, x='day',y='count', ax=ax3)
sns.barplot(data=train, x='hour',y='count', ax=ax4)
sns.barplot(data=train, x='minute',y='count', ax=ax5)
sns.barplot(data=train, x='second',y='count', ax=ax6)

ax1.set(ylabel='Count', title='연도별 대여량')
ax2.set(xlabel='month', title='월별 대여량')
ax3.set(xlabel='day', title='일별 대여량')
ax4.set(xlabel='hour', title='시간별 대여량')
```

- 연도별 대여량은 2011년보다 2012년이 더 많다
- 월별 대여량은 6월에 가장 많고 7 ~ 10월도 대여량이 많다. 그리고 1월에 가장 적다
- 일별대여량은 1일부터 19일까지만 있고, 나머지 날짜는 test.csv에 있다. 그래서 이 데이터는 기준으로 사용하면 안된다.
- 시간 대 대여량을 보면 출퇴근 시간에 대여량이 많은 것 같다. 하지만 주말과 나누어 볼 필요가 있을 것 같다.
- 분, 초 전부 0이므로 의미가 없다.

```python
corrMatt = train[['temp','atemp', 'casual', 'registered', 'humidity', 'windspeed','count']]
corrMatt = corrMatt.corr()
print(corrMatt)

mask =np.array(corrMatt)
mask[np.tril_indices_from(mask)] = False
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(20,10)
sns.heatmap(corrMatt, mask=mask, vmax=.8, square=True, annot=True)
```

- 온도, 습도, 풍속은 거의 연관관계가 없다.
- 대여량과 가장 연관이 높은건 registered로 등록된 대여자가 많지만, test 데이터에는 이 값이 없다.
- atemp와 temp는 0.98로 상관관계가 높지만 온도와 체감온도로 피저를 사용하기에는 적합하지 않을수도 있다.

- 풍속의 경우 0에 숫자가 몰려 있는 것으로 보인다. 아마도 관측되지 않은 수치에 대해 0으로 기록된 것이 아닌가 추측

```python
def concatenate_year_month(datetime):
    return '{0}-{1}'.format(datetime.year, datetime.month)

train['year_month'] = train['datetime'].apply(concatenate_year_month)

print(train.shape)
train[['datetime', 'year_month']].head()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
fig.set_size_inches(18,4)

sns.barplot(data=train, x='year', y='count', ax=ax1)
sns.barplot(data=train, x='month', y='count', ax=ax2)

fig, ax3 = plt.subplots(nrows=1, ncols=1)
fig.set_size_inches(18,4)

sns.barplot(data=train, x='year_month', y='count', ax=ax3)
```

- 2011년보다 2012년의 대여량이 더 많다.
- 겨울보다는 여름에 대여량이 더 많다.
- 2011년과 2012년의 월별 데이터를 이어보면 전체적으로 증가하는 추세이다.
